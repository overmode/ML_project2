```python
# To handdle th
def concac_tup(x):
    x = eval(x)
    s = ""
    for i in range(len(x)):
        s += x[i]

    return s



def construct_features_concat(vocab_filename, tweets, embeddings, nb_concat, relevant_filename):
    """
    tweets : tweets we want to find the features for
    embeddings : vector representations of n-grams
    nb_concat : the number of n-gras we want to concatenate
    relevant_filename : a dictionnary between relevant n-grams and their ratio
    ratios : a dictionanry between n-grams and their ratios
    """
    
    features = []
    invalid_features = [];
    nb_dim = 50

    
    with open(vocab_filename, "rb") as f:
        vocab = pickle.load(f)

    tweets = import_(tweets)
    
    
    # Create the dict to handle the relevant vocab and their ratio


    relevant = pd.read_csv(filepath_or_buffer= relevant_filename, sep="\t", index_col= 1)
    
    # Create the dict to handle all the vocab and their ratio
    #ratios_dict = pd.read_csv(filepath_or_buffer= ratios_filename, sep="\t",index_col=1).to_dict()
    
    
    #Load words from tweet set
    xs = np.load(embeddings)
    
        
    for indl, line in enumerate(tweets):
        
        #we differentiate tweets containing pertinent words, those in dictionnary 'weights'
        count_pertinent = False
        count_other = False
        
        #keep track of best words and their corresponding best ratios
        best_ratios_relevant = [None] * nb_concat
        best_words_relevant = [None] * nb_concat
        
        best_ratios = [None] * nb_concat
        best_words  = [None] * nb_concat
        

        
        for word in extract_tokens(line):
            
            if type(word) == str:
                word = tuple(tokenize(tknzr ,word))

            local_w = vocab.get(word, -1)
            if local_w == -1:
                print("Fuck"+str(word))
            
            if local_w != -1:
                weight = -1
                
                word = str(word)
                
                if(relevant.index.contains(word)):
                    weight = relevant.loc[str(word)]["ratio"]

                    
                if weight != -1:
                    #if the relevant word has one of the nb_concat best ratios seen so far, we add it to best_ratios_relevant
                    #and we forget the replaced old best ratio
                    for ind_r, r in enumerate(best_ratios_relevant) :
                        count_pertinent = True
                        
                        if(r == None or weight > r):
                            
                            #If the word is pertinent, we remember we met it
                            best_ratios_relevant.insert(ind_r, weight)
                            best_words_relevant.insert(ind_r, xs[local_w])
                            best_ratios_relevant = best_ratios_relevant[0:nb_concat]
                            best_words_relevant = best_words_relevant[0:nb_concat]
                            break
                    
                    
                    
                    

                elif not count_pertinent:
                    
                        #we also remember the best ratios of non-relevant words in case there is no relevant word
                        #in the tweet.
                        for ind_r, r in enumerate(best_ratios) :

                            #If the word is not pertinent, we remember we met it
                            count_other = True
                            if(r == None or weight > r):
                                best_ratios.insert(ind_r, weight)
                                best_words.insert(ind_r, xs[local_w])
                                best_ratios = best_ratios[0:nb_concat]
                                best_words = best_words[0:nb_concat]
                                break
                        
                    

            # If we found pertinent words, we only use them
        if(count_pertinent):
            features.append([w for  w in best_words_relevant])

            #if we found only non-pertinent words, we use them anyway
        elif count_other:
            features.append([w for w in best_words])

            #if we did not find words that have representation, we do not try to create features and signal their indices
        else:
            invalid_features.append(indl)

    return features, invalid_features


def accuracy(prediction, actual_emotions):
    return (1 - (np.sum(np.abs(actual_emotions-prediction))/(2*len(actual_emotions)))) * 100


def GloVe(file_name="cooc_partial", destination=embeddings_ts, embedding_dim = 50):
    #load coocurence matrix
    with open(file_name, 'rb') as f:
        cooc = pickle.load(f)    
    
    nmax = 100
    
    eta = 0.001
    alpha = 3 / 4
    epochs = 10
    
    xs = np.random.normal(size=(cooc.shape[0], embedding_dim))
    ys = np.random.normal(size=(cooc.shape[1], embedding_dim))
   
    #Construct vector representations xs for words
    for epoch in range(epochs):
        print("epoch {}".format(epoch))
        loading_counter = 0
        for ix, jy, n in zip(cooc.row, cooc.col, cooc.data):
            
            f = ((n / nmax)**alpha) if n < nmax else 1
            inter_cost = (xs[ix]@(ys[jy]) - np.log(n))
            # We compute the gradients for both context and main vector words
            grad_main = f * inter_cost * ys[jy]
            grad_context = f * inter_cost * xs[ix]
    
            # Update the vector words
            xs[ix] = xs[ix] - (eta * grad_main)
            ys[jy] = ys[jy] - (eta * grad_context)
            
            if loading_counter%20000==1:
                    print("{:.1f}".format(loading_counter/len(cooc.col)*100), "%", end='\r')
            loading_counter+=1
            
    #Store xs in destination file
    np.save(file=destination, arr=xs)
    print("finished")
```

```python
def construct_features(tweets, embeddings, weights, path_vocab_pick = 'global_vocab_cut=5.pkl'):
    features = []
    invalid_features = [];
    nb_dim = 50
    pertinence = 35
    
    # 
    with open(path_vocab_pick, 'rb') as f :
        vocab = pickle.load(f)
    
    # Create the dict to handle the relevant vocab
    relevant = pd.read_csv(filepath_or_buffer= "relevant_vocab_pert=0.3_count=250", sep="\t", index_col= 1)
    dict_relevant = relevant.to_dict()
    
    
    #Load words from tweet set
    xs = np.load(embeddings)

    for indl, line in enumerate(tweets):
        if indl%10000 == 0:
            print("{:.1f}".format(indl/len(tweets)*100), "%", end="\r")
        #we differentiate tweets containing pertinent words, those in dictionnary 'weights'
        sum_w_pertinent = np.zeros(nb_dim)
        sum_w_others = np.zeros(nb_dim)

        count_pertinent = 0
        count_other = 0

        for token in extract_tokens(line):
            local_w = vocab.get(token, -1)
            if local_w != -1:
                
                token = str(token)
                
                if(relevant.index.contains(token)):
                    weight = relevant.loc[str(token)]["ratio"]
                
                weight = dict_relevent.get(token, -1)
                
                if weight != -1:

                    #If the word is pertinent, we add its word representation to others pertinents word's representation
                    count_pertinent += weight*pertinence
                    sum_w_pertinent += xs[local_w] * (weight*pertinence)

                else:

                    #If the word is not pertinent, we add its representation to non-pertinent words representations
                    count_other += 1
                    sum_w_others += xs[local_w]

            # If we found pertinent words, we only use them
        if(count_pertinent != 0):
            features.append(sum_w_pertinent/count_pertinent)

            #if we found only non-pertinent words, we use them anyway
        elif count_other!= 0:
            features.append(sum_w_others/count_other)

            #if we did not find words that have representation, we do not try to create features and signal their indices
        else:
            invalid_features.append(indl)

    invalid_features = np.array(invalid_features)
    features = np.array(features)

    return features, invalid_features
```
